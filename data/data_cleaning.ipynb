```python
# install dependencies
import pandas as pd
import seaborn as sns
```

```python
# load the movies dataset into a pandas dataframe
df = pd.read_csv("original_data.csv")
```

```python
# display the first 5 rows with the head method
df.head()
```

```python
# the info method shows the basic information regarding the dataset
df.info()
```

```python
# the describe method displays the summary statistics for the numerical columns
df.describe()
```

```python
# the isnull.sum method gives the total number of missing values in each column.
# isnull provides the booleans for whether a cell is null or not, and sum counts up all of the nulls
df.isnull().sum()
```

```python
# With the nulls discovered, we will delete rows with N/A values to clean the data a bit to ensure the graphs are not skewed too much.
df.dropna(subset=['rating'], inplace=True)
df.dropna(subset=['released'], inplace=True)
df.dropna(subset=['score'], inplace=True)
df.dropna(subset=['votes'], inplace=True)
df.dropna(subset=['writer'], inplace=True)
df.dropna(subset=['star'], inplace=True)
df.dropna(subset=['country'], inplace=True)
df.dropna(subset=['budget'], inplace=True)
df.dropna(subset=['gross'], inplace=True)
df.dropna(subset=['company'], inplace=True)
df.dropna(subset=['runtime'], inplace=True)
df.isnull().sum()
```

```python
# plotly
# import plotly.plotly as py
from plotly.offline import init_notebook_mode, iplot, plot
import plotly as py
init_notebook_mode(connected=True)
import plotly.graph_objs as go
```

```python
# data preparation

# This prototype experiments with identifying various statistics 
# of scores based off of comedy movies, showcasing an example of how
# the filter function may work when studying different genres
comedyDF = df[df.genre == 'Comedy']

trace0 = go.Box(
    y=comedyDF.score,
    name = 'scores of comedy films',
    marker = dict(
        color = 'rgb(12, 12, 140)',
    )
)
data = [trace0]
iplot(data)
```

```python
# This graph prototype focuses on studying a specific range of
# data, specifically rows that contain movies with the top
# grossing profit made in production based off of a cutoff.

df_topMovies = df.sort_values(by='gross', ascending=False)[:25]

# create trace1 
trace1 = go.Bar(
                x = df_topMovies['name'],
                y = df_topMovies['gross'],
                name = "Top Grossing Movies",
                marker = dict(color = 'rgba(255, 174, 255, 0.5)',
                             line=dict(color='rgb(0,0,0)',width=1.5)),
                text = df_topMovies['genre'] + " " + (df_topMovies['rating']))
data = [trace1]
layout = go.Layout(barmode = "group")
fig = go.Figure(data = data, layout = layout)
iplot(fig)
```
